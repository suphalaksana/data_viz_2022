##**Cleaning Data**

### data1 ตาราง alldb.csv 

```python
data3 =  data1[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey','title']] #เลือกคอลัมม์ที่ต้องการ
data3 #แสดงค่า
```

```python
data3.isnull().sum() #ตรวจดูค่า missing พบว่ายังมีค่า missing อยู่
```

```python
data4 = data3.dropna() #.dropna เพื่อกำจัดค่า missing ออกเนื่องจากข้อมูล Missing มีจำนวนมาก และหากแทนค่าผิดสามารถทำให้ข้อมูลผิดพลาดได้
data4
```

```python
data4.isnull().any() #ตรวจดูค่า missing อีกครั้งพบว่า ไม่มีค่า missing แล้ว
```

**กำหนดค่า User เป็นตัวเลข 5 หลัก โดยใช้คำสั่ง  if**

```python
MS1=list() ## กำหนดให้MSเป็นลิสต์ในการเติมค่าloop
for each_row in data4.iterrows(): ## ทำการวนลูปในแถวของ data4
    if len(str(each_row[1]['user']))== 5 : #สร้างเงื่อนไขให้ user มีจำนวนตัวอักษร 5 ตัว
        if str(each_row[1]['user']) <= '99999' : #ถ้า user มีค่า น้อยกว่า 100000
          MS1.append(True) ## ถ้าตรงตามเงื่อนไขให้เพิ่มเข้าไปใน MS
        elif str(each_row[1]['user']) >= '99999' : #ถ้า user มีค่า มากกว่า 100000
            MS1.append(False) ## ถ้าไม่ตรงตามเงื่อนไขไม่ให้เพิ่มเข้าไปใน MS
    else:
        MS1.append(False) ## ถ้่าไม่ตรงตามเงื่อนไขทั้งหมด ไม่ให้เพิ่มเข้าไปใน MS

data4[MS1] ## แสดงผล
```

```python
data5 =data4[MS1] ##ทำการกำหนดData4 เป็นค่าที่ทำการCleanแล้ว
```

```python
data6 = data5[data5['user']!= '6724-']  ##พบค่าผิดปกติ คือ 6724- จากการเช็คข้อมูล User จึงทำการตัดออก
```

**กำหนดค่า site เป็นตัวเลข 5 หลัก โดยใช้คำสั่ง  if**

```python
MS2=list() ## กำหนดให้MSเป็นลิสต์ในการเติมค่าloop
for each_row in data6.iterrows(): ## ทำการวนลูปในแถวของ data6
    if len(str(each_row[1]['site']))== 5 : #สร้างเงื่อนไขให้ site มีจำนวนตัวอักษร 5 ตัว
        if str(each_row[1]['site']) <= '99999' : #ถ้า site มีค่า น้อยกว่า 100000
          MS2.append(True) ## ถ้าตรงตามเงื่อนไขให้เพิ่มเข้าไปใน MS
        elif str(each_row[1]['site']) >= '99999' : #ถ้า site มีค่า มากกว่า 100000
            MS2.append(False) ## ถ้าไม่ตรงตามเงื่อนไขไม่ให้เพิ่มเข้าไปใน MS
    else:
        MS2.append(False) ## ถ้่าไม่ตรงตามเงื่อนไขทั้งหมด ไม่ให้เพิ่มเข้าไปใน MS

data6[MS2] ## แสดงผล
```

```python
data7 =data6[MS2] ##ทำการกำหนดData 6 เป็นค่าที่ทำการCleanแล้ว
```

```python
data8 = data7[data7['site']!= '043.1']  ##พบค่าผิดปกติ คือ 043.1 จากการเช็คข้อมูล site จึงทำการตัดออก
```
