# Parallel Coordinates

วาดกราฟ parallel coordinate เปรียบเทียบ [อายุ และ เพศ] โดยใช้ class เป็น risks (เลือกมาแค่ 3 risks)

```python
import pandas as pd #package pandas ใช้ในการจัดการข้อมูลที่เป็นตาราง พวกไฟล์ CSV
import os
```

```python
from google.colab import drive  #เชื่อมต่อ google drive กับ colab
drive.mount('/content/drive')
path = '/content/drive/My Drive/dataviz_2021_data'
```

```python
data_file_path = os.path.join(path,'confirmed-cases.csv') 
data_df = pd.read_csv(data_file_path)
```

```python
data_df
```

```python
data_h = data_df[['sex','age','risk']] #เลือกเอาเฉพาะข้อมูลที่ต้องการใช้
data_h
```

```python
data_h.isnull().any() #ตรวจสอบค่าMissing  พบว่า True คือยังมี missing อยู่
```

```python
data_risk = data_h.dropna() #แก้ Missing ด้วยการ Dropna
data_risk  #จากข้อมูล 816989 แถว 3 คอลัมน์ หลัง dropna เหลือ 712916 แถว 3 คอลัมน์
```

```python
set(data_risk['risk'])
```

```python
data_risk.isnull().any() # False คือไม่มีค่า missing แล้ว
```

```python
foreign = data_risk[data_risk['risk']=='State Quarantine'] #เลือกเอา risk เฉพาะ State Quarantine
foreign
```

```python
investigation = data_risk[data_risk['risk']=='สนามมวย'] #เลือกเอา risk เฉพาะสนามมวย
investigation
```

```python
other = data_risk[data_risk['risk']=='อาชีพเสี่ยง เช่น ทำงานในสถานที่แออัด หรือทำงานใกล้ชิดสัมผัสชาวต่างชาติ เป็นต้น'] #เลือกเอาเฉพาะอาชีพเสี่ยง เช่น ทำงานในสถานที่แออัด หรือทำงานใกล้ชิดสัมผัสชาวต่างชาติ เป็นต้น
other
```

```python
data = pd.concat([foreign,investigation,other]) #รวมตาราง
data
```

```python
data_k = data.replace('หญิง',1) #กำหนดเพศหญิงเป็น 1
data_k
```

```python
data_final = data_k.replace('ชาย',100) #กำหนดเพศชายเป็น100
data_final
```

```python
data_y = data_final[['age','risk']] #เลือกเอาเฉพาะข้อมูลที่จะทำกราฟ เลือกข้อมูล age  risk จากตาราง data_final
data_y
```

```python
data_final['risk'].unique() #เลือกดูชื่อrisk
```

```python
pd.plotting.parallel_coordinates(data_final, 'risk',color=['#FF00E0', '#00FFE2', '#ECFF00']); #พอตกราฟ  State Quarantine  , สนามมวย,อื่นๆ
```
