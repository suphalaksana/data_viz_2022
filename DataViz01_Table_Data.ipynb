```python
import pandas as pd  #ข้อมูลลักษณะตาราง
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')  #อ่านไฟล์ในเน้ตได้
df
```

```python
df['Name'].unique()  #ชื่อของดอกไม้มีท้้งหมด 3 พันธุ์
```

# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name')  #ฟังก์ชัน plot graph  คือ pd.plotting
#colum ที่ใช้แบ่งแยก คือ Name  กราฟนี้อ่านว่า ดอก virginica ใหญ่กว่าดอก versicolor
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']) #Sepal  Petal ยาวเท่ากันหมดเพราะมันไม่หยักลงมา
# libary เป็น matplotlib
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']);
#ใส่ colon ; เพื่อให้กราฟสวยขึ้น  ไม่มีตัวแสดงผลออกมา
```

### change color

```python
pd.plotting.parallel_coordinates(df, 'Name',color=['r','y','k']) 
```

```python
pd.plotting.parallel_coordinates(df, 'Name',color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]]) #ใช้สีที่ผสมเอง
```

```python
pd.plotting.parallel_coordinates(df, 'Name',color=['r','y','k']) #ใช้โค้ดสี html
```

## Specify color by colormap

```python
pd.plotting.parallel_coordinates(df, 'Name',colormap='Set1') #ใช้สีที่เข้ากัน ที่มีคนรวมไว้แล้ว
```

# Matplotlib

```python
from matplotlib import pyplot as plt   #จาก libary ใหญ่ import มาแค่ matplotlib   แล้วเอา pyplot มาชื่อ plt
```

##  Heatmap Matriix

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize']=[25,25]  #กำหนดขนาดของ figure
plt.imshow(df.iloc[:,:-1],interpolation='nearest') #ไม่เอาคอลัมน์สุดท้ายมา plot เพราะเป็นตัวอักษร 
# imshow โชว์เป็นรูปภาพ  #ค่ามากสีจะสว่าง  ค่าน้อยสีจะเข้ม
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap='hot') #สี cmap hot #ค่ามากสีจะสว่าง  ค่าน้อยสีจะเข้ม
```

## Sub plot

```python
plt.subplot(1,3,1) # subplot(มีกี่แถว, มีกี่หลัก, ตัวนี้เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')  #เอามาแค่ 50 ตัวไม่เอาแถวสุดท้าย

plt.subplot(1,3,2) # subplot(มีกี่แถว, มีกี่หลัก, ตัวนี้เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2) # subplot(มีกี่แถว, มีกี่หลัก, ตัวนี้เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,100,:-1],interpolation='nearest',cmap='hot')
```
