# confirmed-cases

ข้อมูล confirmed-cases ตั้งแต่วันแรก - 11/08/2021

```python
data_covid1
```

## ตรวจสอบ Missing

```python
data_covid1.isnull().any()
```

จะเห็นได้ว่า ข้อมูล data_covid1  ยังมีค่า Missing อยู่

## แทน missing ด้วยค่าที่เหมาะสม (fillna())

```python
data_covid1_updated = data_covid1.fillna(value={'Notified date':'ไม่มีข้อมูล','sex':'ไม่มีข้อมูล','Unit':'ไม่มีข้อมูล','nationality':'ไม่มีข้อมูล','province_of_isolation':'ไม่มีข้อมูล','risk':'ไม่มีข้อมูล','province_of_onset':'ไม่มีข้อมูล','district_of_onset':'ไม่มีข้อมูล'})  # เลข -99 คือการใสเลขที่ไม่ซ้ำจากอันเดิม  #ตัวแปรชื่อ value  #แทนค่า missing ด้วย 'ไม่รู้'
data_covid1_updated.head()
```

```python
data_covid1_updated.isnull().any()
```

## แทน age ด้วยค่า mean

```python
data_covid_mean1 = data_covid1_updated.fillna(value={'age':data_covid1['age'].mean()})  #เติม missing ด้วยค่าเฉลี่ย
```

```python
data_covid_mean1.isnull().any()  #คอลัมน์ sex ไม่มี missing
```
