หาอายุเฉลี่ยของเพศหญิง

````
this_data = data_covid[['sex','age','province_of_onset']]  #เลือกคอลัมน์ที่จะใช้
````

````
this_data  #print ออกมา
````

````
female = this_data[this_data['sex']=='หญิง']  #เลือกจากตาราง this data เลือกมาแค่ที่เป็นผู้หญิง
female
````

````
female['age']  #ชี้คอลัมน์ที่เป็นอายุของเพศหญิง
````

````
sum = 0  #เริ่มต้นมาเป็น 0 เพราะสร้างตัวแปร sum มารับค่าที่จะเอามาบวกกันทั้งหมด เป็นเลขอื่นไม่ได้เพราะถ้าเอามาบวกกันเรื่อยๆค่ามันจะเพี้ยน(คุณสมบัติการบวกทั่วไป)
N = 0  #นับว่ามีคนทั้งหมดกี่คน เพราะข้อมูลอาจจะมีค่าว่าง
for a in female['age']:  #วนลูปค่าอายุ
    
    if a > 0:  #เช็คว่ามากกว่า 0 คืออายุ คนปกติ ไม่มีข้อมูลผิดพลาด
        sum += a # sum = sum + a  #ถ้า a มากกว่า 0 ก็จะนำค่าใารวมใน  sum
        N += 1  #คำนวนจำนวนคนทั้งหมดใน sum

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง {sum/N}')  #sum/n = อายุเฉลี่ยของผู้ป่วยหญิง
````

````
female_KK = female[female['province_of_onset']=='ขอนแก่น']  #สนใจข้อมูลผู้ป่วยหญิงในจังหวัดขอนแก่น

sum = 0
N = 0
for a in female_KK['age']:#วนลูปในค่าอายุ
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง ในจังหวัดขอนแก่น {sum/N}')
````

### การวนลูป record ในตาราง  (```.iterrows```)

````
this_data = data_covid[['sex','age','province_of_onset']]  #ตัด Data มาจาก Data covid
this_data
````

````
for each_row in this_data.iterrows():   #function iterrows จะสร้าง list ในแต่ละแถวเป็นฟังก์ชันของ pandas เพราะหลัง in ต้องเป็น list
    # print(each_row)  #คือ print แถวที่ 0 
    # print(each_row[1])  #คือ print แถวที่ 1 ที่มีข้อมูล
    # print(each_row[1]['sex'])  #ดึงค่าในแต่ละแถว
    if (each_row[1]['age'] <= 20) and (each_row[1]['province_of_onset'] == 'ขอนแก่น'):  # print แถวที่ age น้อยกว่า เท่ากับ 20   และจังหวัดขอนแก่น
        print(each_row)  # มี 5 คน
````

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] <= 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)  # ถ้า age น้อยกว่า เท่ากับ 20   และจังหวัดขอนแก่น  ให้ print ออกมา
        print(this_data.iloc[each_row,:])   # ได้มาเฉพาะค่าที่เาต้องการ
````
