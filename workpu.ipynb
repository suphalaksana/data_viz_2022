```python
path = '/content/drive/My Drive/dataviz_2021_data'
GW = pd.read_csv(os.path.join(path, 'gw.csv'))
```

```python
GW #เรียกดูข้อมูล
```

##เช็คค่าว่าง

```python
GW.isnull()  #False ไม่มีค่าว่าง
```

```python
GW.isnull().any()#any สรุปข้อมูลเป็นราย column  #true คือเป็นค่า ที่ missing
```

#ลบค่า missing เฉพาะใน column ที่เราสนใจ

```python
GW_nomissing = GW.dropna(subset=['X'])  #missing ลบเฉพาะที่ไม่มี X
```

```python
GW_nomissing
```

```python
GW_nomissing.isnull().any() #any สรุปข้อมูลเป็นราย column  #true คือเป็นค่า ที่ missing
```

```python
GW_nomissing2 = GW_nomissing.dropna(subset=['Y'])  #missing ลบเฉพาะที่ไม่มี X
```

```python
GW_nomissing2
```

```python
GW_nomissing2.isnull().any() #any สรุปข้อมูลเป็นราย column  #true คือเป็นค่า ที่ missing
```

```python
GW_nomissing2['X']  #ชี้คอลัมน์ 
```

```python
GW_nomissing2['Y']  #ชี้คอลัมน์ 
```
