##ต่อแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1) # frac คือสลับแถว อาจจะไม่เหมือนกันทุกคน
covid_data.head()
```

##เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)  #ตารางหลักคือ personal_data
personalcovid_data.head()
```

##Save Table

```python
personalcovid_data.to_csv(os.path.join(path,'my_processed_data.csv'))
```

## List(row oriented)

```python
data = [['Jomes LLC',150,200,140],  # สร้าง list ที่เก็บค่าในแต่ละแถว
        ['Alpha',200,210,215],
        ['Blue',50,90,95]]
headers = ['account','Jan','Feb','Mar'] #กำหนดหัวคอลัมน์ ลำดับใน listt ต้องตรงกัน

df2 = pd.DataFrame.from_records(data,columns=headers)
df2
```
